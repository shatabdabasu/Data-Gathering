```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}
webpage = requests.get('https://books.toscrape.com/?page=1',headers=headers, timeout=10).text
```

```python
webpage
```

```python
soup=BeautifulSoup(webpage, 'lxml')
```

```python
soup
```

```python
soup.find_all('h1')
```

```python
h3_tags = soup.find_all('h3')
for books in h3_tags:
  print(books.text.strip())
```

```python
links = []

for book in h3_tags:
    a = book.find("a")
    if a:
        links.append(a["href"])

print(links)
```

```python
for link in links:
    print(link)
```

Let’s open just ONE link and pull one piece of data. No loops yet.

```python
first_link = links[0]
print(first_link)
```

```python
full_link = "https://books.toscrape.com/" + first_link
full_link
```

```python
r = requests.get(full_link)
book_soup = BeautifulSoup(r.text, "html.parser")
```

```python
book_soup
```

```python
title = book_soup.find("h1").text.strip()
print(title)
```

```python
price = book_soup.find("p", class_="price_color").text.strip()
print(price)
```

```python
price = book_soup.find("p", class_="price_color").text.strip()
print(price)
```

```python
stock = book_soup.find("p", class_="instock availability").text.strip()
print(stock)
```

```python
desc_div = book_soup.find("div", id="product_description")
description = desc_div.find_next("p").text.strip()
print(description)
```

```python
table = book_soup.find("table", class_="table table-striped")
print(table)
```
